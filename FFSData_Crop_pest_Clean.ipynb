### Label Encoding some columns

#### soil_condition

```python
label_mapping = {
    'Dry': 1,
    'Moist (Wapsa)': 2,
    'Wet': 3,
    'Water Logged': 4,
    '1': 1,
    '2': 2,
    '3': 3,
    '4': 4,
}

pest_data['soil_condition'] = pest_data['soil_condition'].map(label_mapping)
pest_data = pest_data.dropna(subset=['soil_condition'])
pest_data['soil_condition'] = pest_data['soil_condition'].astype(int)
pest_data
```

#### wind_condtion

```python
pest_data['wind_conditions'].unique()
pest_data = pest_data[~pest_data['wind_conditions'].isin(['Weed Types & Intensity : 2', 'Weed Types & Intensity : 4'])]
pest_data['wind_conditions'].unique()
label_mapping_wind = {
    'Slow': 1,
    'Calm': 2,
    'Medium': 3,
    'High': 4,
    '1': 1,
    '2': 2,
    '3': 3,
    '4': 4,
     1 : 1,
     2 : 2,
     3 : 3,
     4 : 4,
}

pest_data['wind_conditions'] = pest_data['wind_conditions'].map(label_mapping_wind)
```

#### dropping height, canpoy and pd ratio

```python
pest_data = pest_data.drop(columns=['height', 'canopy','pd_ratio'])
```
